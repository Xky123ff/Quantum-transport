```python
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(0, 0.5, 500)

y = (1 - r**2) / (1 + r + r**2)

plt.figure(figsize=(8, 6))
plt.plot(r, y, label=r'$Y = (1-r^2) / (1+r+r^2)$')

plt.title('Function Plot')
plt.xlabel('r')
plt.ylabel('Y')
plt.xlim(0, 0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python

r = np.linspace(0, 0.5, 500)

y_hall = (1 - r**2) / (1 + r + r**2)
y_long = r / (1 + r + r**2)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

ax1.plot(r, y_hall)
ax1.set_title('rxy')
ax1.set_ylabel('R_xy (a.u.)')
ax1.grid(True)

ax2.plot(r, y_long, color='orange')
ax2.set_title('rxx')
ax2.set_xlabel('r')
ax2.set_ylabel('R_xx (a.u.)')
ax2.grid(True)

plt.xlim(0, 0.5)
plt.tight_layout()
plt.show()
```
